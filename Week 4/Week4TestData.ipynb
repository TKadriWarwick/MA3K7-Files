```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

dim = 7
sampSize = 10_000

def genResults(n: int, m: int) -> list:
    results = []
    for i in range(m):
        size = int(np.floor((n**2)/2))
        matrixA = np.ones(size) #half zeros
        matrixB = np.zeros(size) #half ones
        matrix = np.append(matrixA, matrixB)
        if n%2 == 1:
            matrix = np.append(np.random.randint(low=0, high=2), matrix) #adds a random value if the dimension is odd
        np.random.shuffle(matrix) #shuffles the values in the array
        matrix = np.reshape(matrix, (-1, n)) #makes the matrix square
        results.append(np.linalg.det(matrix)) #records the determinant
        
    return results

outcomes = genResults(dim, sampSize)
refined = Counter(outcomes) #collects and organises the data
nums = list(refined.keys())
occurences = list(refined.values())

plt.bar(nums, occurences, width=0.8)
plt.title("Occurences of determinants with dimension %s" % dim)
plt.xlabel("Determinant")
plt.ylabel("Occurence")
plt.ylim(ymin=0)
plt.show()
```

```python
sampSize = 5_000
dim = []
points = []

for i in range(12):
    i += 2
    dim.append(i)
    temp = genResults(i, sampSize)
    occurences = Counter(temp)[0]
    j = np.divide(occurences,sampSize)
    points.append(j)

plt.scatter(dim, points)
plt.title("Probability of zero determinant against dimension of matrix")
plt.xlabel("Dimension")
plt.ylabel("Probability of zero Determinant")
plt.ylim(ymin=0)
plt.show()
```

Clearly by random input, with the obvious exception of dimension 1, the matrix is most likely to have a zero determinant regardless of the size of the matrix. The closest dimension to the probability of a zero determinant being a half was 7. (Excluding 1.)
