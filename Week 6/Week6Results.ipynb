```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
def hatFunc(size: int) -> int: # Function that mimics the hat game and returns the final value
    nums = [(i+1) for i in range(size)]
    np.random.shuffle(nums) # Shuffles the numbers
    
    for i in range(size -1):
        temp = abs(nums[0] - nums[1])
        nums = nums[2:] # Takes and removes the first two
        nums.insert(np.random.randint(0, len(nums) + 1), temp) # Inserts the new value randomly
        
    return nums[0]
```

```python
hatSize = 2024
samples = 500

results = []
for i in range(samples):
    results.append(hatFunc(hatSize))

refined = Counter(results) #collects and organises the data
sortedrefined = dict(sorted(refined.items()))
nums = list(sortedrefined.keys())
occurences = list(sortedrefined.values())

myPlot = plt.boxplot(sortedrefined, vert=False)
plt.title("Results of running the hat function %s times with value %s" % (samples, hatSize))
plt.xlim(xmax= hatSize)
plt.yticks([])
plt.show()

median = myPlot['medians'][0].get_xdata()[1]
quartiles = [myPlot['boxes'][0].get_xdata()[1], myPlot['boxes'][0].get_xdata()[2]]

print("Smallest value: " + str(nums[0]) + ", with occurence: " + str(occurences[0]))
print("First quartile: " + str(quartiles[0]))
print("Median:         " + str(median))
print("Third quartile: " + str(quartiles[1]))
print("Largest value:  " + str(nums[-1]))
```
